```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
df=pd.read_csv(r"C:\Users\AGM MSME PNB\OneDrive\Desktop\Sashidhar\Data scince\Projects\Final Project\PROJECT-2 (RATINGS OF GUVI COURSES)\3.1-data-sheet-guvi-courses.csv")
df
```

```python
null_rows = df[df.isnull().any(axis=1)]
```

```python
null_rows
```

```python
df = df.dropna()
```

```python
df['datetime'] = pd.to_datetime(df['published_timestamp'])
df['year'] = df['datetime'].dt.year
df
```

```python
#course_id has not significance in prediction
# url not have significane and mostly url and course_title are common
# published_timestamp and datetime are removed as we are taking the year column for rating
columns_to_drop = ['course_id', 'url','published_timestamp','datetime']
df = df.drop(columns_to_drop, axis=1)
df
```

```python
ordinal_cols=['level']
nominal_cols = ['course_title','subject']

le=LabelEncoder()
for col in ordinal_cols:
    le.fit(df[col])
    df[col]=le.transform(df[col])
    
dff = pd.get_dummies(df[nominal_cols])
dff
```

```python
dff[['price',  'num_subscribers',  'num_reviews', 'num_lectures', 'level', 'Rating','content_duration','year']] = df[['price',  'num_subscribers',  'num_reviews', 'num_lectures', 'level', 'Rating','content_duration','year']]
```

```python
dff = dff.astype('int')
dff
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

model = LinearRegression()

X=dff.drop('Rating',axis=1)
y=dff['Rating']
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
lr=LinearRegression()
lr.fit(x_train,y_train)
y_train_pred=lr.predict(x_train)
y_test_pred=lr.predict(x_test)
y_test_pred
```

```python
from sklearn.metrics import mean_squared_error,r2_score
import math 

print("Train error" , mean_squared_error(y_train,y_train_pred))

print("Test error" , mean_squared_error(y_test,y_test_pred))

print(" Root of Train error" , math.sqrt(mean_squared_error(y_train,y_train_pred)))
print("Root of Test error",math.sqrt(mean_squared_error(y_test,y_test_pred)))

```
